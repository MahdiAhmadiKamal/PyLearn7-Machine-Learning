```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
data = pd.read_csv('Boston.csv')
data.head()
```

```python
data.corr()
```

```python
plt.scatter(data['nox'], data['age'])
plt.show()
```

```python
plt.scatter(data['age'], data['lstat'])
plt.show()
```

```python
X = np.array((data['nox'], data['age'])).T
Y = np.array(data['lstat'])
```

```python
X.shape
```

```python
Y.shape
```

Splitting dataset to train and test datasets:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=None)
```

Using my object oriented linear least squares (LLS) class:

```python
from lls import LLS
```

```python
lls = LLS()
```

```python
slp = lls.slope(X_train, y_train)
slp
```

```python
x1, x2 = np.meshgrid(X_train[:,0], X_train[:,1])
plane = slp[0] * x1 + slp[1] * x2 


fig = plt.figure( dpi=200, facecolor='w', edgecolor='k')
ax = plt.axes(projection='3d', )
ax.plot_surface(x1, x2, plane , alpha= 0.3, label="y_pred")
plt.scatter(X_train[:,0] , X_train[:,1] , y_train , label="data" )
plt.legend()
plt.show()
```
