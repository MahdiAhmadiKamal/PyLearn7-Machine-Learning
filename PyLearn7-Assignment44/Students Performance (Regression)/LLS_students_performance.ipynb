```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

Generating continuous random dataset for the problem:

```python
X, y, coef = datasets.make_regression(n_samples=200,n_features=1,noise=30,coef=True , random_state=20) 

X = np.interp(X, (X.min(), X.max()), (0.5, 6))
y = np.interp(y, (y.min(), y.max()), (0, 20))
```

```python
plt.scatter(X,y,marker = "o"  , c=X ,  cmap = 'jet')
plt.legend()
plt.xlabel('Study hours')
plt.ylabel('Grade')
plt.title('Students grades per study hours')
```

Splitting dataset to train and test datasets:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)
```

Using my object oriented linear least squares (LLS) class:

```python
from lls import LLS
```

```python
lls = LLS()
```

```python
slp = lls.slope(X_train, y_train)
slp
```

```python
plt.scatter(X_train, y_train, label='Data')
plt.plot(X_train, slp * X_train, color='red', label='Linear Regression')
plt.legend()
plt.show()
```

Using linregress from scipy library:

```python
from scipy.stats import linregress
```

```python
slop, intercept, rvalue, pvalue, stderr =linregress(X_train.T, y_train)
```

```python
slop
```

```python
plt.scatter(X_train, y_train, label='Data')
plt.plot(X_train, slop * X_train, color='green', label='Linear Regression')
plt.legend()
plt.show()
```

Comparison:

```python
plt.scatter(X_train, y_train, label='Data')
plt.plot(X_train, slp * X_train, color='red', label='Linear Regression (class LLS)')
plt.plot(X_train, slop * X_train, color='green', label='Linear Regression (scipy)')
plt.xlabel('Study hours')
plt.ylabel('Grade')
plt.title('Students grades per study hours')
plt.legend()
plt.show()
```
